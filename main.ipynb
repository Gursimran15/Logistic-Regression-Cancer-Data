## Logistic Regression

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
# z = np.dot(X, theta)
# h = sigmoid(z)
# def loss(h, y):
#     return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
# gradient = np.dot(X.T, (h - y)) / y.shape[0]
# lr = 0.01
# theta -= lr * gradient
def Logistic_Regression(X,Y,e,lr):
    X2= X_val.T
    X=X.T
    Y=Y.reshape(1,Y.shape[0])
    y=y_val.values
    Y2= y.reshape(1,y.shape[0])
    epochs=e
    learningrate=lr
    losstrack=[]
    losstrack2 =[]
    m2 = X2.shape[1]
    m = X.shape[1]
    w = np.random.randn(X.shape[0], 1)*0.01
    b = 0 
    for epoch in range(epochs):
        z = np.dot(w.T, X) + b
        p = sigmoid(z)
        cost = -np.sum(np.multiply(np.log(p), Y) + np.multiply((1 - Y), np.log(1 - p)))/float(m)
        print(cost)
        losstrack.append(np.squeeze(cost))
        z2 = np.dot(w.T, X2) + b
        p2 = sigmoid(z2)
        cost2 = -np.sum(np.multiply(np.log(p2), Y2) + np.multiply((1 - Y2), np.log(1 - p2)))/float(m2)
        losstrack2.append(np.squeeze(cost2))
        dz = p-Y
        dw = (1 / float(m)) * np.dot(X, dz.T)
        db = (1 / float(m)) * np.sum(dz)
        w = w - learningrate * dw
        b = b - learningrate * db 
    plt.plot(losstrack,color="red",label="Train Loss")
    plt.plot(losstrack2,color="purple",label="Validation Loss")
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.legend()
    return w,b
```
